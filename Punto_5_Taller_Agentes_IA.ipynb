```
from IPython.display import display
import ipywidgets as widgets
import time
import random

class Tablero:
  def __init__(self, tamanoCelda=(40, 40), nCeldas=(5,5), predeterminado =[]):
    self.out = widgets.HTML()
    display(self.out)
    self.tamanoCelda = tamanoCelda
    self.nCeldas = nCeldas
    self.casilla = []
    self.predeterminado = predeterminado
    aux = 0
    for x in range(0, nCeldas[0]):
      for y in range(0, nCeldas[1]):
        if aux in self.predeterminado:
          self.casilla.append(Casilla(x,y, aux, valor="ðŸ¥Ÿ"))
        else:
          self.casilla.append(Casilla(x,y, aux))
        aux += 1

  def dibujar(self, agente):
    tablero = "<table border='1'>{}</table>"
    filas = ""
    aux = 0
    for i in range(self.nCeldas[0]):
      s = ""
      for j in range(self.nCeldas[1]):
        if agente.x == j and agente.y == i:
         if self.casilla[aux].valor == "ðŸ¥Ÿ":
            contenido = f"<div style='transform: rotate({agente.angulo}deg);font-size:{agente.tamanoEmoticon}px;'>ðŸ™Š</div>"
            self.casilla[aux].valor = ""
            self.predeterminado.remove(self.casilla[aux].numero)
         else:
            contenido = f"<div style='transform: rotate({agente.angulo}deg);font-size:{agente.tamanoEmoticon}px;'>{agente.emoticon}</div>"
        elif self.casilla[aux].valor == "ðŸ¥Ÿ":
            contenido = f"<div style='transform: rotate({agente.angulo}deg);font-size:{agente.tamanoEmoticon}px;'>ðŸ¥Ÿ</div>"
        else:
          contenido =""
        aux += 1

        s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".\
        format(alto=self.tamanoCelda[0], ancho=self.tamanoCelda[1], 
               contenido=contenido)
      filas += "<tr>{}</tr>".format(s)
    tablero = tablero.format(filas)
    self.out.value=tablero

  def validar(self, x, y):
      posicion = self.nCeldas[0]*y+x;
      if self.casilla[posicion].valor == "ðŸ¥Ÿ":
          return True
      else:
          return False

```

```
class Casilla:
  def __init__(self, x, y, numero, valor=""):
    self.x = x
    self.y = y
    self.numero = numero
    self.valor = valor
  
  def __str__(self):
    return "Numero "+str(self.numero)+", x :"+ str(self.x)+ ", y :"+ str(self.y)+", valor= "+self.valor
  
  def valor(self):
    print(self.x, self.y)
```

```
class Agente:
  
  def __init__(self, x=0, y=0, angulo=0, emoticon="ðŸ™‰", tamanoEmoticon=25):
    self.x = x
    self.y = y
    self.angulo = angulo
    self.emoticon = emoticon
    self.tamanoEmoticon = tamanoEmoticon
    self.puntaje = 0
   

  def moverseAbajo(self, escenario):
    if escenario.predeterminado != []:
      self.y += 1
      self.puntaje += 1

  def moverseArriba(self, escenario):
    if escenario.predeterminado != []:
      self.y -= 1
      self.puntaje += 1

  def moverseDerecha(self, escenario):
    if escenario.predeterminado != []:
      self.x +=1
      self.puntaje += 1
  
  def moverseIzquierda(self, escenario):
    if escenario.predeterminado != []:
      self.x-=1
      self.puntaje += 1

  def revisar(self, escenario, parte):
    contador, contador2 = 0, 0

    if self.x+1 < escenario.nCeldas[0]:
      if escenario.validar(self.x+1, self.y):
        escenario.dibujar(self)
        self.moverseDerecha(escenario)
        time.sleep(1)
        escenario.dibujar(self)
        self.moverseIzquierda(escenario)
        time.sleep(1)
        escenario.dibujar(self)
        
    if self.x-1 >= 0 and escenario.predeterminado != []:
      if escenario.validar(self.x-1, self.y):
        escenario.dibujar(self)
        self.moverseIzquierda(escenario)
        time.sleep(1)
        escenario.dibujar(self)
        self.moverseDerecha(escenario)
        time.sleep(1)
        escenario.dibujar(self)
        
    if self.y-1 >= 0 and escenario.predeterminado != []:
      if escenario.validar(self.x, self.y-1):
        escenario.dibujar(self)
        self.moverseArriba(escenario)
        time.sleep(1)
        escenario.dibujar(self)  
        contador2=1; 
        
    if self.y+1 < escenario.nCeldas[0] and escenario.predeterminado != []:
      if escenario.validar(self.x, self.y+1):
        escenario.dibujar(self)
        self.moverseAbajo(escenario)
        time.sleep(1)
        escenario.dibujar(self)
        contador=1;
    
    if contador == 0 and parte == 1 and escenario.predeterminado != []:
      if self.y+1 < escenario.nCeldas[0]:
          self.moverseAbajo(escenario)
          time.sleep(1)
          escenario.dibujar(self)
         
    if contador2 == 0 and parte == 2 and escenario.predeterminado != []:
      if self.y-1 >= 0:
          self.moverseArriba(escenario)
          time.sleep(1)
          escenario.dibujar(self)
          


  
```

```
class CicloDeLimpieza:
  def __init__(self):
    self.agente = Agente()
    while True:
      dimension = int(input("Â¿QuÃ© dimensiÃ³n desea? Minimo 3 -> "))
      if dimension >= 3: break
    
    while True:
      empanadas = int(input("Â¿CuÃ¡nta suciedad quiere colocar? -> "))
      if empanadas <= dimension*dimension-1: 
        break
      print("Cantidad invalida, debe ser menor a dimension^2")

    aleatorio = int(input("Â¿Desea ubicar la suciedad aleatoriamente? Si=1 || No=2 -> "))
    suciedad = []
    if aleatorio == 2:
      while True:
        if empanadas < dimension**2: break
      print("")
      for i in range(empanadas):
        while True:
          y = int(input("PosiciÃ³n en x: "))
          x = int(input("PosiciÃ³n en y: "))
          if x >= 0 and x < dimension and y >= 0 and y < dimension: break
          else:
            print("-----Datos fuera de rango-----")  
        suciedad.append((x,y))
    else:
      suciedad = random.sample(range(1,dimension*dimension), empanadas)
    self.escenario = Tablero(nCeldas=(dimension,dimension), predeterminado = suciedad);
    self.iniciarCiclo()
    print("La aspiradora ha realizado su trabajo satisfactoriamente");
    print("Medida de rendimiento: ",self.agente.puntaje)

  def bajar(self):
    aux=0
    while aux < self.escenario.nCeldas[0]:
      self.escenario.dibujar(self.agente)
      self.agente.revisar(self.escenario,1)
      aux+=1

  def derecha (self, cantidad):
    for x in range(0,cantidad):
      self.agente.moverseDerecha(self.escenario)
      time.sleep(1)
      self.escenario.dibujar(self.agente)

  def subir(self):
    aux=0;
    while aux <= self.escenario.nCeldas[0]:
      self.escenario.dibujar(self.agente)
      self.agente.revisar(self.escenario,2)
      aux+=1;

  def iniciarCiclo(self):
    total = self.escenario.nCeldas[0]*2;
    while total < (self.escenario.nCeldas[0] * self.escenario.nCeldas[0]) and self.escenario.predeterminado != []:
      self.bajar()
      if self.agente.x+3 < self.escenario.nCeldas[0] and self.escenario.predeterminado != []:
        self.derecha(3)
      elif self.agente.x+2 < self.escenario.nCeldas[0] and self.escenario.predeterminado != []:
        self.derecha(2)
      else: break

      
      self.subir();
      total += 15
      if self.agente.x+3 < self.escenario.nCeldas[0] and self.escenario.predeterminado != []:
        self.derecha(3)
      elif self.agente.x+2 < self.escenario.nCeldas[0] and self.escenario.predeterminado != []:
        self.derecha(2)
      else: break
  


```
